```python
#inheritance

class Parent:
    def __init__(self,p1,p2):
        self.p1=p1
        self.p2=p2
    
    def parent_property(self):
        print("Parent property details")
        print("Parent property 1: ",self.p1)
        print("Parent property 2: ",self.p2)
        
    
class Child(Parent):
    def __init__(self,c1,c2,p1,p2):
        super().__init__(p1,p2)
        self.c1=c1
        self.c2=c2
        
    def child_property(self):
        print("child property details")
        print("child property 1: ",self.c1)
        print("child property 2: ",self.c2)
        

        
c1=Child("bike",89000,"House","Volvo")
c1.parent_property()
c1.child_property()
        
        
```

```python
#multiple inheritance

class Parent1:
    def set_parent1_property(self,p1,p2):
        self.p1=p1
        self.p2=p2
    
    def show_parent1_property(self):
        print("\n...........................")
        print("Parent1 property details")
        print("Parent1 property : ",self.p1)
        print("Parent1 property : ",self.p2)

        
class Parent2:
    def set_parent2_property(self,p3,p4):
        self.p3=p3
        self.p4=p4
    
    def show_parent2_property(self):
        print("\n...........................")
        print("Parent2 property details")
        print("Parent2 property : ",self.p3)
        print("Parent2 property : ",self.p4)

        
        
    
class Child(Parent1,Parent2):
    def __init__(self,c1,c2):
        
        self.c1=c1
        self.c2=c2
        
    def child_property(self):
        print("...........................")
        print("child property details")
        print("child property 1: ",self.c1)
        print("child property 2: ",self.c2)
        

c1=Child("bike",85000)
c1.child_property()

c1.set_parent1_property("house1",200000)
c1.show_parent1_property()

c1.set_parent2_property("house2",400000)
c1.show_parent2_property()
```

```python
#multilevel inheritance
class Parent:
    def set_parent_data(self,pd):
        self.pd=pd
        
    def show_parent_data(self):
        print("parent data : ",self.pd)
            
class Child(Parent):
    def set_child_data(self,cd):
        self.cd=cd
        
    def show_child_data(self):
        print("Child data : ",self.cd)
           
class Grand_child(Child):
    def set_grandchild_data(self,gcd):
        self.gcd=gcd
        
    def show_grandchild_data(self):
        print("Grand Child data : ",self.gcd)

        
g1=Grand_child()

g1.set_parent_data("house")
g1.show_parent_data()

g1.set_child_data("car")
g1.show_child_data()

g1.set_grandchild_data("bike")
g1.show_grandchild_data()
```

```python
# method overloading is not possible in python

class Myclass:
    def sum(self,a=None,b=None,c=None):
        if a!=None and b!=None and c!=None:
            s=a+b+c
            
        
        elif a!=None and b!=None:
                s=a+b
                
        return s
    
    
m1=Myclass()
print(m1.sum(10,20,30))
print(m1.sum(10,20))

print(m1.sum("asd","vvv"))
```

```python
#method overriding

class Myclass1:
    def res(self,a,b):
        print("Addition : ",a+b)
        
    def sub(self,a,b):
        print("Subtraction : ",a-b)

class Myclass2(Myclass1):
    def res(self,a,b):
        super().res(10,2)
        print("Multiplication : ",a*b)

obj=Myclass2()
obj.res(10,2)


```

```python
#abstraction

from abc import ABC,abstractmethod

class Father(ABC):
    @abstractmethod
    def disp(self):
        pass
    
    def show(self):
        print("Concret method")
        
class child(Father):
    def disp(self):
        print("Child class")
        print("Defining abstract method")
        
c=child()
c.disp()
c.show()
```

```python
from abc import ABC,abstractmethod

class vehicle(ABC):
    
    @abstractmethod
    def start(self):
        pass
    
    def use(self):
        print("Going from one location to another location")
    
class car(vehicle):
    def start(self):
        print("Start with key")
    

class bike(vehicle):
    def start(self):
        print("Start with kick")
        
c1=car()
c1.start()
c1.use()

b1=bike()
b1.start()
b1.use()
```

```python
#----------------------------multithreading in python

from time import sleep
from threading import *

class Hello(Thread):
    def run(self):
        for i in range(0,5):
            print("hello")
            sleep(1)
            

class Hii(Thread):
    def run(self):
        for i in range(0,5):
            print("hii")
            sleep(1)
            

t1=Hello()
t2=Hii()

t1.start()

sleep(0.2)

t2.start()

t1.join()
t2.join()

print("bye")
```

```python
#DECORATOR
# decorator function is a function that accept function as a prameter and return a function
# -inhancing particular function without modification in actual function


    
    
def decor(num):
    def inner():
        print("before inhancing")
        num()
        print("After inhancing function")
        
    return inner     
        

@decor    
def num():
    print("hello world")
    print("hello akshay")
    

num()    
```

```python
# nested class

# writing class inside another class


class Army:                                    #outer class
    def __init__(self):
        self.name='Rahul'
        self.g=self.gun()                     #creating inner class object
    
    def show(self):
        print("Name is ",self.name)
        
    class gun:                                #inner class
        def __init__(self):
            self.name='Ak47'
            self.capacity='75 rounds'
            self.length='34.3 in'
            
        def show(self):
            print("Gun name",self.name)
            print("Gun capacity",self.capacity)
            print("Gun length",self.length)
            
a1=Army()                                    #creating outer class object
a1.show()

g1=Army().gun()                              #creating inner class another object outside class
g1.show()

print(a1.g.name)                             #accessing inner class data 
```
